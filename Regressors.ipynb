```python
import pandas as pd
import numpy as np
import seaborn as sns
from tqdm.notebook import tqdm
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("sdg_index_2000-2022_TARGET=nextyear.csv",)
```

```python
data = data.dropna()
```

```python
data
```

```python
data.iloc[:, 4:-1]
```

```python
X = data.iloc[:, 4:-1]
Y = data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error
```

```python
regr.feature_importances_
```

```python
r2s, maes, rmses = [], [], []
coefs_matrix = []
for i in tqdm(range(1, 101)):
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=i)
    regr = RandomForestRegressor(max_depth=10, random_state=0)
    regr.fit(X_train, y_train)
    y_pred = regr.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    r2s.append(r2)
    maes.append(mean_absolute_error(y_test, y_pred))
    rmses.append(root_mean_squared_error(y_test, y_pred))
    coefs_matrix.append(regr.feature_importances_)
```

```python
np.mean(r2s)
```

```python
coefs_matrix = np.array(coefs_matrix)
df_coefs = pd.DataFrame(coefs_matrix, columns=X_train.columns)
df_coefs
```

```python
fig, ax = plt.subplots(figsize=(13,5))
b = sns.boxplot(df_coefs)
ax.tick_params(axis='x', rotation=45)
```
