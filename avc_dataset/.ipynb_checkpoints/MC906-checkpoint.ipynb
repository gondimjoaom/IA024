```python
# All libraries used on this notebook must be imported
# in this cell only

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from sklearn.metrics import confusion_matrix
```

```python
# First you must upload the dataset on collab with the name "data_set.csv"

df = pd.read_csv('data_set.csv')

# Dropping columns that we don't want to consider for this project
df.drop(columns=[
    'Diabetes_012',
    'HighBP',
    'HighChol',
    'CholCheck',
    'HeartDiseaseorAttack',
    'AnyHealthcare',
    'NoDocbcCost'],
    inplace=True
        )

df.info()
```

```python
# Strongly unbalanced target variable
target_count = df.Stroke.value_counts()
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sb.countplot(x='Stroke', data=df, ax=ax[0])
ax[1].pie(
    target_count,
    autopct=lambda p:f'{p:.2f}%',
    explode=(0, 0.1),
    labels=target_count.index
    )
plt.show()
```

```python
df = df.dropna()
X = df.drop('Stroke', axis=1)
y = df['Stroke']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42, shuffle=False)
train_df = pd.concat([X_train, y_train], axis=1)

# trying a rebalance
min_class = train_df[train_df['Stroke'] == 1]
maj_class = train_df[train_df['Stroke'] < 1]

min_class_upsampled = resample(min_class, random_state=42, n_samples=int(0.5 * len(train_df)), replace=True)
train_df_upsampled = pd.concat([maj_class, min_class_upsampled])

X_train_up = train_df_upsampled.drop('Stroke', axis=1)
y_train_up = train_df_upsampled['Stroke']
# import Random Forest classifier

from sklearn.ensemble import RandomForestClassifier


# instantiate the classifier

rfc = RandomForestClassifier(random_state=0, n_estimators=50, max_depth=15)



# fit the model

rfc.fit(X_train_up, y_train_up)



# Predict the Test set results

y_pred = rfc.predict(X_test)

# Check accuracy score

from sklearn.metrics import accuracy_score

print('Model accuracy score with 10 decision-trees : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))

confusion_matrix(y_test, y_pred)
```
