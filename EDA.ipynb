```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data1 = pd.read_csv('sdg_index_2000-2022.csv')
```

```python
data1["year"].unique()
```

```python
def add_X_year(row, x):
    current_year = row['year']
    country = row['country']
    if data1[(data1["country"]==country) & (data1["year"]==current_year+x)].shape[0] != 0:
        # print(data[(data["country"]==country) & (data["year"]==current_year+1)]["year"].values[0])
        sdg_index_score_next_year = data1[(data1["country"]==country) & (data1["year"]==current_year+x)]["sdg_index_score"].values[0]
        return sdg_index_score_next_year
    else:
        return pd.NA
```

```python
anos = 10
data1[f'sdg_index_score_next_{anos}_years'] = data1.apply(lambda row: add_X_year(row, x=10), axis=1)
data1.to_csv(f"sdg_index_2000-2022_TARGET=next{anos}years.csv", index=False)
```

```python
from scipy.stats import spearmanr
```

```python
new_data1 = data1.dropna()
```

```python
spearman_feats = []
g_target = new_data1["sdg_index_score_next_year"]
for feat in new_data1.iloc[:, 4:-1].columns:
    g_feat = new_data1[feat]
    rho, pvalue = spearmanr(g_target, g_feat)
    if rho > .5:
        print(feat, rho)
```

```python
sns.boxplot(data=new_data1, y='sdg_index_score_next_year')
```

```python
new_data1["sdg_index_score_next_year_discrete"] = pd.qcut(new_data1["sdg_index_score_next_year"], 4,
                                                          labels = ["baixo", "médio", "médio-alto", "alto"])
```
